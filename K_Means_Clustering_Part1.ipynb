**K-Means:**

K-Means its a problem solving technique. The basic idea of K-means/clustering remains the same i.e reduce the intra-cluster distance and increase the inter cluster distance. In K-means, K is the number of clusters and it a hyper-parameter. 
Let us understand now the intuition first then we will formulate the mathamatical objective function/Optimazation problem.
We first randomly select K initial centroids. A centroids is defined as the  central point of each cluster. Suppose we are selecting k=3 then we will randomly initilise 3 points as centroids of those 3 clusters. 
Then each point is then assigned to the closest centroids. Each collection of of points assigned to centroids forms a cluster.
The centroids are recalutated based on the newly formed clusters and we repeat this process of assignment and update untils no point changes. Or the centroids remain the same.



**Basic K-means Algorithim:**

1: Select K points as initial centroids.

2: repeat

    3: Form K clusters by assigning each point to its closest centroid.
    4: Recompute the centroid of each cluster.
5: until Centroids do not change.

# K-Mean++ :
By improving the initialization techniques, we can get rid of this problem.K-mean ++ use a smart initialization . K-mean++ pick the first centroid randomly then calculate the distance of all points from that centroids. Let us take our previous example:
image.png

After calculating the distance, it will select the next centroid probabilistically such that the probability of choosing a point is propostional to its distance. That means higher distance, higher is the probabilty of that point to be selected as next centroid.
In our example, the probability of the following point (80,91),(85,70)(71,80) for becoming next centroid is much higher than the other points.
Intuitively we are selecting centroids which are further away from each other.This is not done determinisitically and is done propabilistically to avoid effect of outlier.
