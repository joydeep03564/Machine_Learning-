```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import BatchNormalization
```

```
#Loading the dataset
path="/content/drive/My Drive/diabetes.csv"
dataset=pd.read_csv(path)
```

```
dataset.head()

dataset.count()
```

About the data, the dataset contains 8 features like Pregnancies,Glucose,Bloodpressure,SkinThickness,Insulin,BMI,DiabatesPedigreeFunction,age.
The label is a categorirical feature which is either giving 0 or 1.
1 means diabetic 
0 means non diabetic

```
# Spliting the dataset into features and labels 
#dataset.shape
x=dataset.iloc[:,:-1].values
#print(x)
#print(x)
y=dataset.iloc[:,8].values
#print(y)

```

```
Scaler=MinMaxScaler()
Scaler.fit(x)
normalised_x=Scaler.transform(x)
```

```
#Spliting the dataset into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(normalised_x,y,test_size=0.30,random_state=1)

```

```
#Here why only fit is used in X-train, is it due same as training the data by using fit function on the train data 
#and using that fit data we training the test data

# from sklearn.preprocessing import StandardScaler
# scaler=StandardScaler()
# scaler.fit(x_train)
# x_train=scaler.transform(x_train)
# x_test=scaler.transform(x_test)
```

Now our data is preprocessed and ready for building model. We will use Tf.Keras to build the model. Below are the steps of building the model
1. **Define the Model**(In this step we define our model, like what is the size of input layer, How many hiddenlayer and their number of nodes,what will be the activation function like RELU or sigmoid)
Model can be defined in two ways using Sequential API and Functional API
In this notebook, we will do it through sequential API of TF.Keras. It is like stacking the layers one after one.

```
x_train.shape[1]
```

From the above X_train.shape we know that our input layer is 8 dimention.
We will be using RELU as our activation unit and Sigmoid in the last layer.
For this problem, we will choose the initial structure as follows:
inputs_layer(8)->first hidden layer(12)->second_hidden layer(8)->output(1) layer

```
# define the Model

model=Sequential()
model.add(Dense(512,activation='relu',kernel_initializer='he_normal',input_shape=(8,)))
model.add(Dropout(0.5))
model.add(BatchNormalization())
model.add(Dense(64,activation='relu',kernel_initializer='he_normal'))
model.add(Dropout(0.3))
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
```

```
# Compile the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
#Fit the model
model.fit(x_train,y_train,epochs=500,batch_size=32,verbose=2)
```

```
#evaluate the model
train_loss,train_acc=model.evaluate(x_train,y_train,verbose=2)
print(train_loss)
print(train_acc)
```

```
test_loss,test_acc=model.evaluate(x_test,y_test,verbose=2)
```

```
print(f"Test_accurancy:{test_acc}")
print(f"Test_Loss:{test_loss}")
```
